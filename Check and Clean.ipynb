## Want to see about approve comment and remove comment, also take out comments from mods that talk about rules

```python
import pandas as pd
import numpy as np

import os
from dotenv import load_dotenv
load_dotenv()

CLIENT_ID = os.getenv('CLIENT_ID')
CLIENT_SECRET=os.getenv('CLIENT_SECRET')
APP_NAME=os.getenv('APP_NAME')
REDDIT_USERNAME=os.getenv('REDDIT_USERNAME')
REDDIT_PASSWORD=os.getenv('REDDIT_PASSWORD')

import praw
import pandas as pd
import datetime as dt

reddit = praw.Reddit(client_id=CLIENT_ID, client_secret=CLIENT_SECRET, user_agent=APP_NAME, username=REDDIT_USERNAME, password=REDDIT_PASSWORD)
```

```python
comments = pd.read_csv('files/compiled_comments.csv')
comments
```

```python
substring = "**/r/NeutralNews is a curated space.**"
mod = comments[comments['content'].str.contains(substring, regex = False, na = False)].index
comments.drop(mod, inplace = True)
comments
```

```python
approved = comments[comments['action'] == 'approvecomment']
removed = comments[comments['action'] == 'removecomment']
approved
```

```python
print(len(approved))
print(len(removed))
```

```python
for index, approve in approved.iterrows():
    approve_text = approve['content']
    if removed['content'].str.contains(approve_text, regex = False, na = False).any():
        comments.drop(index, inplace = True)
        
        
comments
```
