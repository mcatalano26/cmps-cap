```python
swearwords_df = pd.read_csv("files/swear-words.csv")
swearwords = swearwords_df.a55hole.tolist()

allowed_words = ['gay', 'sexual', 'crack', 'sperm', 'stupid', 'damn', 'prostitute', 'sex', 'kill', 'crack', 
                'oral', 'rape', 'raped', 'screw', 'sleazy', 'murder', 'organ', 'slave', 'strip', 'butt', 'naked',
                'porn', 'bone', 'nazi', 'racy', 'bodily', 'screw', 'pot', 'weed', 'pedophile', 'rapist', 'drunk', 
                'commie', 'condom', 'thrust', 'pedophilia', 'enlargement', 'cocaine', 'beer', 'stroke', 'opium',
                'unwed', 'meth', 'napalm', 'perversion', 'queers', 'sadism', 'sadist', 'seduce', 'seaman', 'seamen']
for word in allowed_words:
    if word in swearwords:
        swearwords.remove(word)

swearwords_df = pd.DataFrame(swearwords, columns = ['swear'])
swearwords_df.to_csv('files/edited-swear-words.csv', index=False)
```

```python
comments_df = pd.read_csv("validation/Validation Comments - Sheet1.csv")
comments_df.head()
```

```python
profanity = []
for index, row in comments_df.iterrows():
    #print(index)
    if pd.notnull(row['Comment text']):
        words = row['Comment text'].split()
    else:
        words = ["fine"]
    prof_check = False
    for word in words:
        if word in swearwords:
            prof_check = True
            #print(word)
    profanity.append(prof_check)

#print(profanity)
#comments_df.iloc[1].content
```
