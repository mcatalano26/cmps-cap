```python
import pandas as pd
```

```python
df = pd.read_csv("files/compiled_comments_2_22_2021.csv")
```

```python
df
```

```python
def exclaim_func(comments_df):
    exclaim_lst = []
    for index, com in comments_df.iterrows():
        try:
            if '!' in com['content']:
                exclaim_lst.append(True)
            else:
                exclaim_lst.append(False)
        except:
            exclaim_lst.append(False)
    return exclaim_lst

df['contains_!'] = exclaim_func(df)
df
```

```python
df.to_csv("files/compiled_comments_2_23_2021", index=False)
```
